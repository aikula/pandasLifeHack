```python
import openpyxl
import pandas as pd
```

```python
wb = openpyxl.load_workbook('test_form.xlsx', data_only=True) #название файла
wbSheet = wb['Лист1'] #название листа
```

```python
# самый простой вариант, не учитывает наличия названия столбцов 
df = pd.DataFrame(wbSheet.values)
df
```

```python
# чуть сложнее вариант, берет первую строку как название столбца
data = wbSheet.values
cols = next(data)[0:] #выталкивает первую строку для названия столбцов
df = pd.DataFrame(data, columns=cols)
df
```
